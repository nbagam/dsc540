```python
## columns whcih has NAN or NULL
df.columns[df.isnull().any()]
```

```python
#Drop columns with Missing values
df1 = df.dropna(axis=1)
print(df1)
# only 2 cloumns left out of 123 columns, with out missing any data
```

```python
# Drop rows with missing values with threshold 50
df2 = df.dropna(axis=0,how ='all',thresh=50)
print(df2)
df2.shape
```

# Chapter 8

```python
# Creating  pivot table
pivot = pd.pivot_table(
    data=df2,
    index='Your gender:',
    aggfunc='count',
    values={' [100 Grand Bar]',' [Black Jacks]'}
    
)
```

```python
print(pivot)
```

```python
# Creating  pivot table
pivot2 = pd.pivot_table(
    data=df2,
    index='Your gender:',
    aggfunc='count',
    values={' [Any full-sized candy bar]',' [Anonymous brown globs that come in black and orange wrappers]',' [Bonkers (the candy)]'}
    
)
```

```python
print(pivot2)
```

```python
# Merge Datasets
pd.merge(pivot,pivot2, left_index=True, right_on='Your gender:')
```
