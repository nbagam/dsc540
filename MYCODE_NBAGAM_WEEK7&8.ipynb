```python
#BeautifulSoup for HTML parsing:
contents = response.content.decode(response.encoding)
soup = BeautifulSoup(contents,'html.parser')
```

```python
## Find all the href tags and store them in the list of links

#Define empty list
lst_links=[]

for link in soup.find_all('a'):
    lst_links.append(link.get('href'))
#print first 30 links
lst_links[:30]
```

```python
# regular expression to numeric digits in the links

# define empty list
booknum=[]  # hold file number for top 100 books
```

```python
#Numbers 19 to 118 in original list of links have the top 100 eBooks numbers.

for i in range(19,119):
    link=lst_links[i]
    link=link.strip()
    n=re.findall('[0-9]+',link)
    if len(n)==1:
        booknum.append(int(n[0]))
        print("\n The file numbers for the top 100 ebooks on gutenberg website\n"+"-"*70)
        print(booknum)
```

```python
# Import requireed libraries
import urllib.request, urllib.parse, urllib.error
import json
```

```python
#Load API key
with open('APIkeys.json') as f:
    keys = json.load(f)
    omdbapi = keys['OMDBapi']
    
```

```python
# Store OMDB URL in variable

serviceurl="http://www.omdbapi.com/?"
```

```python
# create API key variable to pass it to URL
apikey = '&apikey='+omdbapi
```

```python
#Funtion to print the data from the json file
def print_json(json_data):
    list_keys=['Title', 'Year','Rated', 'Released','Runtime','Genre','Director','Writer','Actors','Plot','Language','Country','Awards','Rating', 'Metascore', 'imdbRating','imdbvotes','imdbID']
    print("-"*50)
    for k in list_keys:
        if k in list(json_data.keys()):
            print(f"{k}:{json_data[k]}")
            print("-"*50)
```

```python
# function to downlaod a poster of the movie
def save_poster(json_data):
    import os
    title = json_data['Title']
    poster_url = json_data['Poster']
    #split poster URL by "."
    poster_file_extension=poster_url.split('.')[-1]
    # Read image file
    poster_data = urllib.request.urlopen(poster_url).read()
    savelocation=os.getcwd()+'\\'+'Posters'+'\\'
    # check if the save directory exist
    if not os.path.isdir(savelocation):
        os.mkdir(savelocation)
        filename=savelocation+str(title)+'.'+poster_file_extension
        f=open(filename,'wb')
        f.write(poster_data)
        f.close()
```

```python
# Function to search a movie
def search_movie(title):
    try:
        url = serviceurl + urllib.parse.urlencode({'t':str(title)})+apikey
        print(f'Retreving the data of "{title}" now ...')
        print(url)
        uh = urllib.request.urlopen(url)
        data = uh.read()
        json_data = json.loads(data)
        if json_data['Response']=='True':
            print_json(json_data)
            if json_data['Poster'] !='N/A':
                save_poster(json_data)
        else:
            print("Error encountered:",json_data['Error'])
                
    except urllib.error.URLError as e:
        print(f"ERROR:{e.reason}")
                
```

```python
# test Serach Movie function
search_movie("Titanic")
```

```python
# Search movie using "Random_error"
search_movie("Random_error")
```
